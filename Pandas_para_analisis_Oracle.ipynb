#Analisis de nuevas variables

  - Valor bruto = alquiler + mantenimiento + impuesto
  - Valor por metro cuadrado
  - Valor bruto por metro cuadrado

```python
#Creacion de la columna y datos de valor bruto
datos_residencial['Valor Bruto']= (datos_residencial['Valor']) + (datos_residencial['Mantenimiento']) + (datos_residencial['Impuesto'])
datos_residencial.head(10)
```

```python
#Creacion de la columna y datos de valor por m2
datos_residencial['Valor m2']= (datos_residencial['Valor']) /(datos_residencial['Area'])
datos_residencial['Valor m2'] = datos_residencial['Valor m2'].round(2)
datos_residencial.head(10)
```

```python
#creacion columna valor bruto por m2
datos_residencial['Valor Bruto m2']= ((datos_residencial['Valor Bruto']) /(datos_residencial['Area'])).round(2)
datos_residencial.head(10)
```

```python
#si el tipo es casa debe ser clasificado como casa, sino será departamento
datos_residencial['Tipo Agrupado'] = datos_residencial['Tipo'].apply(lambda x: 'Casa' if 'Casa' in x else 'Departamento')
datos_residencial.head(10)
```

```python
#excluir variables del dataframe
datos_aux = pd.DataFrame(datos_residencial[['Tipo Agrupado', 'Valor m2','Valor Bruto', 'Valor Bruto m2']])
datos_aux.head(10)
```

```python
#1er metodo
del datos_aux['Valor Bruto']
datos_aux.head(10)
```

```python
#2do metodo .pop() muestra los datos eliminados
datos_aux.pop('Valor Bruto m2')
datos_aux.head(10)
```

```python
#3er metodo + de una columna
datos_residencial.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace= True) #debo indicar que son columnas lo que se elimina
datos_residencial.head(10)
```

#Exclusion de Outliers

```python
plt.rc('figure',figsize=(14,6))
```

```python
datos_residencial.boxplot(['Valor'])
```

```python
datos_residencial[datos_residencial['Valor']>=500000]
```

```python
valor=datos_residencial['Valor']
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3-Q1
limite_inferior= Q1-1.5*IIQ
limite_superior= Q3+1.5*IIQ
```

```python
seleccion = ((valor>=limite_inferior) & (valor<=limite_superior))
```

```python
datos_corregido= datos_residencial[seleccion]
```

```python
datos_corregido.boxplot(['Valor'])
```

```python
datos_residencial.hist(['Valor'])
```

```python
datos_corregido.hist(['Valor'])
```

```python
datos_corregido.boxplot(['Valor'],by=['Tipo'])
```

```python
grupo_tipo = datos_residencial.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
#indices por cada tipo de vivienda
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3-Q1
limite_inferior= Q1-1.5*IIQ
limite_superior= Q3+1.5*IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
datos_corregido=pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo=datos_residencial['Tipo']==tipo
  eh_dentro_limite=(datos_residencial['Valor']>=limite_inferior[tipo])&(datos_residencial['Valor']<=limite_superior[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion=datos_residencial[seleccion]
  datos_corregido=pd.concat([datos_corregido,datos_seleccion])

datos_corregido
```

```python
datos_corregido.boxplot(['Valor'],by=['Tipo'])
```

```python
datos_corregido.to_csv('alquiler_residencial_sin_outliers.csv', sep=';', index=False)
```

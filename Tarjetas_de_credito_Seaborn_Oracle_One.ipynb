```python
#rango_edad
def rango_edad(edad):
  if edad>50:
    return('+50')
  elif edad>40:
    return('40-50')
  elif edad>30:
    return('30-40')
  else:
    return('20-30')

tarjetas['rango_edad']=  tarjetas['edad'].apply(lambda x: rango_edad(x))
tarjetas.head()
```

#Test de hipÃ³tesis

```python
sns.lmplot(x='iu',y='valor_factura', data=tarjetas,col='moroso');
```

Linea positiva en ambos casos, pero pendiente diferente, quiere decir que en ambos casos es directamente proporcional la relacion de ambas variables, se debe evaluar si es estadisticamente significativo o no

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribucion de los grupos morosos y no morosos **es la misma**

H<sup>alt<sup>

La distribucion de los grupos morosos y no morosos **no es la misma**

```python
moroso = tarjetas.query('moroso ==1').valor_factura
```

```python
no_moroso = tarjetas.query('moroso ==0').valor_factura
```

```python
#compararemos la distribucion en ambos grupos
resultado=ranksums(moroso,no_moroso)
```

```python
#el p value menor a 0 quiere decir que demostraria la hipotesis alternativa
print(f'El p_value del test es de {resultado.pvalue}')
```
